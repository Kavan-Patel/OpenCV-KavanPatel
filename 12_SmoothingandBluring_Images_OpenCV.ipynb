```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
```

```python
img=cv.imread('opencv-logo.png',1)
img=cv.cvtColor(img, cv.COLOR_BGR2RGB)

kernel= np.ones((5,5), np.uint8)/25

dst= cv.filter2D(img,-1,kernel)     #bluring image and smooothing
blur= cv.blur(img,(5,5))            #bluring image and smooothing
gblur= cv.GaussianBlur(img,(5,5),0)     #bluring less remove noise
median= cv.medianBlur(img, 5)          # remove sault paper noise
bilifilter=cv.bilateralFilter(img, 9, 75,75)     #remove noise by keeping edge 


images=[img, dst, blur, gblur, median, bilifilter]
titles=['Logo',"Filter","Blur","gblur","median","bilifilter"]


plt.figure(figsize=(15,15))
for i in range(6):
    plt.subplot(2,3,i+1), plt.imshow(images[i],'gray')
    plt.xticks([]), plt.yticks([])
    plt.title(titles[i])

plt.show()
```
