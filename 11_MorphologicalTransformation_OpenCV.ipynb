```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
# from matplotlib import interactive
# interactive(True)
# %matplotlib
```

```python
img=cv.imread('smarties.png',0)
_,msk=cv.threshold(img,220, 255,cv.THRESH_BINARY_INV)

kernal= np.ones((4,4),np.uint8)
dilation= cv.dilate(msk,kernal,iterations=3)
erotion= cv.erode(msk,kernal,iterations=2)
opening= cv.morphologyEx(msk, cv.MORPH_OPEN, kernal)
closing= cv.morphologyEx(msk, cv.MORPH_CLOSE, kernal)
mg= cv.morphologyEx(msk, cv.MORPH_GRADIENT, kernal)
th= cv.morphologyEx(msk, cv.MORPH_TOPHAT, kernal)

images=[img, msk, dilation,erotion,opening, closing, mg, th]
titles=['Original', "Mask","dilation","Erotion","Opening","Closing","mg","th"]

plt.figure(figsize=(15,15))
for i in range(8):
    plt.subplot(2,4,i+1), plt.imshow(images[i],'gray')
    plt.xticks([]), plt.yticks([])
    plt.title(titles[i])

plt.show()
```
