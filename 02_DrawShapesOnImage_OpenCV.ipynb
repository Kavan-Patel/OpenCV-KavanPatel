# Learned Methods
1. plt.imshow() shows image in RGB formate and imread reads imag in BGR formate so we need to convert BGR2RGB with cv.cvtColor
2. cv.shape() shape= circle, rectangle, etc...
3. np.zeros()


# Imports

```python
import cv2 as cv
from matplotlib import pyplot as plt
import numpy as np
```

# Variables

```python
img= cv.imread('lena.jpg', 1)
# img= np.zeros([1000,1000], np.uint8)
rgbImage= cv.cvtColor(img,cv.COLOR_BGR2RGB)

rgbImage= cv.line(rgbImage, (0,0), (500,500), (255,0,0), 5)
rgbImage= cv.arrowedLine(rgbImage, (0,500), (490,50), (0,0,255), 5)
rgbImage= cv.rectangle(rgbImage, (425,50), (50,500), (0,255,0), 5)
font= cv.FONT_HERSHEY_COMPLEX
rgbImage= cv.putText(rgbImage, 'Lena', (50,500),font, 2, (255,255,255),4)

plt.imshow(rgbImage)
plt.show()
```
